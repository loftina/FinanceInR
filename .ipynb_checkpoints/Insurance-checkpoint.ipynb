The cost of an insurance claim for a bodily injury due to a car accident follows a normal distribution with mu 50 and sigma of 20. The number of claims follows a poisson distribution with lambda 2 per month. Find the expected cost to the insurance company for each month in the next year.

```r
months <- seq(1, 12)
average_claims_per_month <- 10
average_cost_per_claim <- 10
standard_deviation_of_cost <- 10
```

```r
claims_per_month = rep(0, 12)
for (month in months){
    claims_per_month[month] <- rpois(1, month*2)
}
claims_per_month
```

```r
cost_to_company = rep(0, 12)
month_counter <- 1
max_claim_so_far = -1
for (claims in claims_per_month){
    if (claims > max_claim_so_far){
        max_claim_so_far = claims
        for (claim in 1:claims){
            cost_of_claim <- rnorm(1, 0, 1)
            cost_to_company[month_counter] <- cost_to_company[month_counter] + cost_of_claim
        }
    }
    else{
        cost_to_company[month_counter] <- cost_to_company[month_counter-1]
    }
    month_counter <- month_counter + 1
}
```

```r
for (month in months) {
    cat('Expected cost for month', month, 'is', cost_to_company[month], '\n')
}
```

```r
length(months)
length(cost_to_company)
plot(months, cost_to_company, type = 's')
lines(c(0,12), c(0,0), col = 'red')
```

```r
annual_cost <- sum(cost_to_company)
cat('Expected annual cost is', annual_cost)
```

```r
cat('Simple expected annual cost', length(months)*average_claims_per_month*average_cost_per_claim)
```
