```r
x <- seq(1, 11)
y <- seq(from = 0, to = 100, by = 10)
for (i in 1:length(y)) {
    y[i] <- y[i] + rnorm(1, 0, 5)
}
length(x)
length(y)
mean.y <- mean(y)
model1 <- lm(formula = y ~ x)
```

```r
plot(x = x, y = y)
abline(h = mean.y, col = 'green') # null hypothesis
abline(model1, col = 'red')
```
