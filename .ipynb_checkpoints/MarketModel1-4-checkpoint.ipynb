```r
market_model1 = function (lambda, beta, time = seq(1,252)) {
    
    jump_at_t = rep(0, length(time))
    for (t in time){
        jump_at_t[t] <- rpois(1, lambda * t)
    }
    plot(x = time, y = jump_at_t, type = 'l')
    
    movement <- rep(0,length(time))
    max <- -1
    counter <- 1
    for (jump in jump_at_t){
        if (jump > max) {
            max <- jump
            movement[counter] <- sum(rnorm(jump)) 
        }
        else {
            movement[counter] <- movement[counter - 1]
        }
        counter <- counter + 1
    }
    
    model <- movement + beta * time
    return(model)
}
```

```r
plot(x = seq(1, 252), y = market_model1(23, 0), type = 'l')
```

```r
# lambda determines size of jumps, beta determines growth, mu determines mean
# time determines time interval
market_model3 = function(lambda, beta, mu, time = seq(1, 252)) {
    jump_at_t = rep(0, length(time))
    for (t in 1:length(time)){
        jump_at_t[t] <- rpois(1, lambda) # removed t
    }
    plot(x= time, y= jump_at_t, type = 'l')
    movement <- rep(0,length(time))
    counter <- 1
    for (jump in jump_at_t){
        x <- rnorm(jump)
        movement[counter] <- sum(x)
        counter <- counter + 1
    }
    
    movement <- movement + beta * sqrt(time) + mu # squared time and added 750
    return (movement)
}
```

```r
time = seq(1,252)
model = market_model3(lambda = 2, beta = 1/2, mu = 750)
plot(x = time, y = model, type = 'l')
lines(time, cumsum(model)/1:length(model), type = 'l', col = 'red')
```

```r
market_model4 = function (lambda, beta, sigma, time = seq(1,252)) {
    
    jump_at_t = rep(0, length(time))
    for (t in time){
        jump_at_t[t] <- rpois(1, lambda)
    }
    
    movement <- rep(0,length(time))
    counter <- 1
    for (jump in jump_at_t){
        movement[counter] <- sum(rnorm(jump, mean = counter, sd = sigma)) 
        counter <- counter + 1
    }
    
    model <- movement + beta * time
    return(model)
}
```

```r
plot(x = time, y = market_model4(lambda = 23,beta =  0, sigma = 0), type = 'l')
```
