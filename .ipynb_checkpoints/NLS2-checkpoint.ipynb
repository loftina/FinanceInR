```r
time = seq(1,252)
#lambda avg(stock)
lambda <- .25
jump_at_t = rep(0, 252)
max <- -1
for (t in 1:length(time)){
    if (jump_at_t[t] > max) {
        max <- jump_at_t[t]
        jump_at_t[t] <- rpois(30, lambda * t) # removed t
    }
}
plot(x = time, y = jump_at_t, type = 's')
movement <- rep(0,252)
counter <- 1
for (jump in jump_at_t){
    if (jump > max) {
        max <- jump
        x <- rnorm(jump)
        movement[counter] <- sum(x) 
    }
    else {
        movement[counter] <- movement[counter - 1] + rnorm(n = 1, sd = 2) # added normal var
    }
    counter <- counter + 1
}
#beta mean
beta <- .25
movement <- movement + beta * sqrt(time) + 750 # squared time and added 750
plot(x = time, y  = movement, type = 'l')
lines(x = time, y = cumsum(movement)/1:length(movement), col = 'red')
```
