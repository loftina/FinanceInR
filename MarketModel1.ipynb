```r
time = seq(1,252)
```

```r
#lambda avg(stock)
lambda <- 23
jump_at_t = rep(0, 252)
for (t in time){
    jump_at_t[t] <- rpois(1, lambda * t)
}
plot(x = time, y = jump_at_t, type = 's')
```

```r
movement <- rep(0,252)
max <- -1
counter <- 1
for (jump in jump_at_t){
    if (jump > max) {
        max <- jump
        x <- rnorm(jump)
        movement[counter] <- sum(x) 
    }
    else {
        movement[counter] <- movement[counter - 1]
    }
    counter <- counter + 1
}
```

```r
#beta mean
beta <- 1/2
movement <- movement + beta * time
plot(x = time, y  = movement, type = 'l')
```
